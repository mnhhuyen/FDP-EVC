```python
import json
import pandas as pd

with open("hirisplex_results_FN.json") as f:
    data = json.load(f)

rows = []

for item in data:
    if "input_csv" not in item:
        print("Skipping sample missing input_csv:", item.get("sample", item))
        continue
    
    row = {
        "sample": item["sample"],
        "input_csv": item["input_csv"]
    }
    
    # Flatten result list with indexes
    for idx, r in enumerate(item["result"]):
        row[f"result/{idx}/trait"] = r["trait"]
        row[f"result/{idx}/p_value"] = r["p_value"]
        row[f"result/{idx}/auc_loss"] = r["auc_loss"]
    
    rows.append(row)

df = pd.DataFrame(rows)
df.to_csv("hirisplex_results_flat.csv", index=False)

print("Saved hirisplex_results_flat.csv")
print("Total rows saved:", len(df))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# ============================
# 1. Load data
# ============================
df = pd.read_csv("hirisplex_results_FN_v2.csv")
df = df.dropna(subset=["input_csv"])
df.shape
```

```python
# ============================
# 2. Parse SNP dosage vector
# ============================
def parse_snp_vector(x):
    # x dạng string chứa 2 dòng (header SNP + vector SNP)
    # ví dụ: "rs...,rs...,rs...\n0,1,2,..."
    lines = x.strip().split("\n")
    snp_values = lines[1].split(",")
    return np.array(snp_values, dtype=int)

df["snp_vector"] = df["input_csv"].apply(parse_snp_vector)

df["snp_vector"]
```

```python
X = np.stack(df["snp_vector"].values)

print("Feature shape:", X.shape)
```

```python
X[2]
```

```python
# ============================
# 3. Extract probabilities
# ============================
# mỗi sample có 14 trait * 3 cột (trait, p_value, auc_loss)
# ta chỉ cần trait + p_value 14 nhóm: 0..13 để chọn ra label

NUM_TRAITS = 14

traits = df[[f"result/{i}/trait" for i in range(NUM_TRAITS)]].values
probs = df[[f"result/{i}/p_value" for i in range(NUM_TRAITS)]].values.astype(float)

# ============================
# 4. Lấy label bằng argmax
# ============================

# Eye: indices 0,1,2
eye_probs = probs[:, 0:3]
eye_label = np.argmax(eye_probs, axis=1)

# Hair: indices 3,4,5,6,7,8
hair_probs = probs[:, 3:9]
hair_label = np.argmax(hair_probs, axis=1)

# Skin: indices 9..13
skin_probs = probs[:, 9:14]
skin_label = np.argmax(skin_probs, axis=1)
```

```python
import numpy as np

print("Eye:", np.unique(eye_label, return_counts=True))
print("Hair:", np.unique(hair_label, return_counts=True))
print("Skin:", np.unique(skin_label, return_counts=True))
```

```python
from sklearn.model_selection import train_test_split

X_train_val, X_test, y_eye_train_val, y_eye_test, y_hair_train_val, y_hair_test, y_skin_train_val, y_skin_test = train_test_split(
    X, eye_label, hair_label, skin_label,
    test_size=0.2, random_state=42, stratify=hair_label
)

print("Train+Val:", X_train_val.shape, "Test:", X_test.shape)
```

```python
import pandas as pd
import numpy as np

# Train full
train_full_df = pd.DataFrame(X_train_val)
train_full_df["eye"] = y_eye_train_val
train_full_df["hair"] = y_hair_train_val
train_full_df["skin"] = y_skin_train_val

# Test
test_df = pd.DataFrame(X_test)
test_df["eye"] = y_eye_test
test_df["hair"] = y_hair_test
test_df["skin"] = y_skin_test
```

```python
from sklearn.model_selection import StratifiedKFold
import os
import numpy as np
import pandas as pd

# Lấy features & labels
X_full = train_full_df.iloc[:, :-3].values  # 41 SNP columns
y_eye = train_full_df["eye"].values
y_hair = train_full_df["hair"].values
y_skin = train_full_df["skin"].values
```

```python
k = 5
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)

os.makedirs("folds", exist_ok=True)

fold_idx = 0

for train_index, valid_index in skf.split(X_full, y_hair):  # stratify theo hair
    fold_dir = f"folds/fold_{fold_idx}"
    os.makedirs(fold_dir, exist_ok=True)

    train_df = train_full_df.iloc[train_index].reset_index(drop=True)
    valid_df = train_full_df.iloc[valid_index].reset_index(drop=True)

    train_df.to_csv(f"{fold_dir}/train.csv", index=False)
    valid_df.to_csv(f"{fold_dir}/valid.csv", index=False)

    print(f"Saved fold {fold_idx}: train={train_df.shape}, valid={valid_df.shape}")
    fold_idx += 1
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
import pandas as pd
import numpy as np

tasks = ["eye", "hair", "skin"]
models = ["LR", "SVM", "RF"]

# Lưu kết quả dạng: results[task][model] = list accuracy 5 fold
results = {task: {m: [] for m in models} for task in tasks}

for fold in range(5):
    print(f"\n=== Fold {fold} ===")
    
    # Load train/valid
    train_df = pd.read_csv(f"folds/fold_{fold}/train.csv")
    valid_df = pd.read_csv(f"folds/fold_{fold}/valid.csv")

    # Split X, y
    X_train = train_df.iloc[:, :-3].values
    X_valid = valid_df.iloc[:, :-3].values
    
    for task in tasks:
        y_train = train_df[task].values
        y_valid = valid_df[task].values
        
        # Train LR
        clf_lr = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)
        clf_lr.fit(X_train, y_train)
        acc_lr = f1_score(y_valid, clf_lr.predict(X_valid), average="macro")
        results[task]["LR"].append(acc_lr)
        
        # Train SVM
        clf_svm = SVC(class_weight="balanced", kernel="rbf")
        clf_svm.fit(X_train, y_train)
        acc_svm = f1_score(y_valid, clf_svm.predict(X_valid), average="macro")
        results[task]["SVM"].append(acc_svm)
        
        # Train RF
        clf_rf = RandomForestClassifier(n_estimators=300, class_weight="balanced", random_state=42, n_jobs=-1)
        clf_rf.fit(X_train, y_train)
        acc_rf = f1_score(y_valid, clf_rf.predict(X_valid), average="macro")
        results[task]["RF"].append(acc_rf)
        
        print(f"{task.upper()} → LR={acc_lr:.4f}, SVM={acc_svm:.4f}, RF={acc_rf:.4f}")
```

```python
print("\n=== AVERAGE VALIDATION ACCURACY (5 folds) ===")
summary = {}

for task in tasks:
    summary[task] = {m: np.mean(results[task][m]) for m in models}

summary_df = pd.DataFrame(summary)
print(summary_df)
```

```python
best_models = {task: max(summary[task], key=summary[task].get) for task in tasks}
print("=== BEST MODEL PER TASK ===")
print(best_models)
```
